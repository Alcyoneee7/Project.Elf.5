```python
#-----------------------------------------     1
#Import File
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import csv
#-----------------------------------------      1


#-----------------------------------------2
#Url , Change Every Projects
#ONLY URL,
driver = webdriver.Chrome()
url= 'https://worldpopulationreview.com/country-rankings/religion-by-country'
driver.maximize_window()
driver.get(url)

time.sleep(5)
content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")

#Change Only (Div,Class)
Container = soup.findAll("tr")

with open('World Religion.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f , delimiter ="|")
    header = ['Country', 'Christ', 'Muslim',"Atheist", "Hindu", "Buddhist" ,'Folk', 'Other',"Jews" ] 
    writer.writerow(header)

for p in Container:
    nv = p.findAll('td')
    
    Country = nv[0].text if len(nv) > 1 else ''
    Christ = nv[1].text if len(nv) > 1 else ''
    
    Muslim = nv[2].text if len(nv) > 1 else ''
    Atheist = nv[3].text if len(nv) > 1 else ''
    Hindu = nv[4].text if len(nv) > 1 else ''
    
    Buddhist = nv[5].text if len(nv) > 1 else ''
    Folk = nv[6].text if len(nv) > 1 else ''
    Other = nv[7].text if len(nv) > 1 else ''
    
    Jews = nv[8].text if len(nv) > 1 else ''

        
        
        
    print(Country)
    print(Christ)
    
    print(Muslim)
    print(Atheist)
    print(Hindu)
    
    print(Buddhist)
    print(Folk)
    print(Other)
    print(Jews)
    
    bliss = [Country, Christ, Muslim,Atheist, Hindu, Buddhist ,Folk, Other,Jews ]    
    
    with open('World Religion.csv', 'a', newline='', encoding='UTF8') as f:
        writer = csv.writer(f , delimiter ="|")
        writer.writerow(bliss)
    
driver.quit()
```
